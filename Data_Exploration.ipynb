# Importing Libraries

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline


import seaborn as sns
```

```python
Data = pd.read_csv('Pressure_with_Anomaly_2.csv')
```

## Converting Timestamp to Datetime

```python
Data['datetime'] = pd.to_datetime(Data["timestamp"])
```

```python
Data.head()
```

# Time Series of Value by DateTime

```python
Data.plot(x= 'datetime', y= 'value', figsize = (12,10), fontsize = 20)
plt.xlabel('DateTime', fontsize = 12)
plt.ylabel('Value', fontsize = 12)
plt.title('Time Series of Value by DateTime', fontsize = 20)
```

# Setting Datetime as Index

```python
Data = Data.set_index('datetime')
```

```python
Data.drop('timestamp', axis = 1, inplace = True)
```

```python
Data.head()
```

# Minimum Yearly Pressure Value 

```python
Data.resample(rule='A').min()
```

```python
Data.resample(rule='A').min().plot(kind='line', figsize=( 12 , 6 ), fontsize =20)
plt.title('Minimum Yearly Pressure Value', fontsize = 20)
plt.grid()
```

# Maximum Yearly Pressure Values 

```python
Data.resample(rule='A').max()
```

```python
Data.resample(rule='A').max().plot(kind = 'line', figsize = (12,6), fontsize = 20)
plt.title('Maximum Yearly Pressure Value', fontsize = 20)
plt.grid()
```

# Average Yearly Pressure Value 

```python
Data.resample(rule='A').mean()
```

```python
Data.resample(rule='A').mean().plot(kind = 'line', figsize = (12,6), fontsize = 20)
plt.title('Average Yearly Pressure Value', fontsize = 20)
plt.grid()
```

# Maximum Monthly Pressure Values

```python
Data.resample(rule='M').max().plot(kind = 'barh', figsize = (12,10), fontsize = 20)
plt.title('Maximum Monthly Pressure Values', fontsize = 20)
plt.grid()
```

```python
Data.resample(rule='M').max().plot(kind = 'hist', figsize = (12,10), fontsize = 20)
plt.title('FREQUENCY DISTRIBUTION OF MAX PRESSURE VALUES', fontsize = 20)
plt.grid()
```

# Minimum Monthly Pressure Values

```python
Data.resample(rule='M').min().plot(kind = 'line', figsize = (12,10), fontsize = 20)
plt.title('MINIMUM PRESSURE VALUES', fontsize = 20)
plt.grid
```

```python
Data.resample(rule='M').min().plot(kind = 'hist', figsize = (12,6), fontsize = 20)
plt.title('DISTRIBUTION OF MINIMUM PRESSURE VALUES', fontsize = 20)
plt.grid()
```

# Average Monthly Pressure

```python
Data.resample(rule='M').mean()
```

```python
Data.resample(rule='M').mean().plot(kind = 'line', figsize = (12,10), fontsize = 20)
plt.title('AVERAGE MONTHLY PRESSURE', fontsize = 20)
plt.grid()
```
