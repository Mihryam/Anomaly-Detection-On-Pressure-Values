```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
%matplotlib inline


import seaborn as sns
```

```python
Data = pd.read_csv('Pressure_with_Anomaly_2.csv')
```

## Converting Timestamp to Datetime

```python
Data['datetime'] = pd.to_datetime(Data["timestamp"])
```

```python
Data.head()
```

# Time Series of Value by DateTime

```python
sns.set_style("darkgrid")
Data.plot.line( y='value', figsize=( 12 , 8 ))
```

```python
Data = Data.set_index('datetime')
```

```python
Data.head()
```

# Minimum Yearly Pressure Value 

```python
Data.resample(rule='A').min()
```

```python
Data.resample(rule='A').min().plot(kind='line', figsize=( 12 , 6 ))
```

# Maximum Yearly Pressure Values 

```python
Data.resample(rule='A').max()
```

```python
Data.resample(rule='A').max().plot(kind = 'line', figsize = (12,6))
```

# Average Yearly Pressure Value 

```python
Data.resample(rule='A').mean()
```

```python
Data.resample(rule='A').mean().plot(kind = 'line', figsize = (12,6))
```

# Maximum Monthly Pressure Values

```python
Data.resample(rule='M').max().plot(kind = 'barh', figsize = (12,10))
```

# Minimum Monthly Pressure Values

```python
Data.resample(rule='M').min().plot(kind = 'line', figsize = (12,10))
```

# Average Monthly Pressure

```python
Data.resample(rule='M').mean()
```

```python
Data.resample(rule='M').mean().plot(kind = 'line', figsize = (12,10))
```

```python
Data.resample(rule='T').max().plot(kind = 'bar', figsize = (12,10))
```
